```python
import pytrends
from pytrends.request import TrendReq
pytrends = TrendReq(hl="en-US", tz=360)
pytrends.build_payload(kw_list = ["Salman Khan"])
df = pytrends.interest_over_time()
df["Salman Khan"].plot(figsize = (20, 5), color = 'black', linewidth = 3)
```
